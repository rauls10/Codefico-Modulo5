# **Atividades Práticas: Eventos em JavaScript, HTML, e CSS**

---


### **Introdução:**
> **Estou disponibilizando um arquivo de template para a realização da atividade**

> Caso queiram explorar mais o assunto ou tenham dificuldades, deixo esse site de recomendação: https://www.w3schools.com/js/js_events_examples.asp

### Nesta atividade, você terá a oportunidade de aplicar os conceitos de eventos em JavaScript, HTML e CSS. Utilize o template fornecido (index.html, styles.css e script.js) como base para realizar as seguintes tarefas.

# **1. Onclick:**

> Ao clicar no botão "Clique-me!", queremos exibir uma mensagem simples.\
>**Passo a passo:**
>1.  Abra o arquivo script.js.
>2.  Selecione o botão usando document.getElementById.
>3.  Adicione um evento de clique (onclick) ao botão.
>4.  No manipulador de eventos, exiba um alerta simples.

# **2. Onchange:**
> Quando o texto no campo de entrada for alterado, queremos mudar a cor do texto.\
>**Passo a passo:**
>1.  Selecione o campo de entrada usando document.getElementById.
>2.  Adicione um evento de mudança (onchange) ao campo.
No manipulador de eventos, mude a cor do texto.

# **3. Oninput:**
> Queremos que a cor do texto mude à medida que o usuário digita.\
>**Passo a passo:**
>1.  Selecione o campo de entrada usando document.getElementById.
>2.  Adicione um evento de entrada (oninput) ao campo.
>3.  No manipulador de eventos, mude a cor do texto enquanto o usuário digita.

# **4. Onkeydown, Onkeypress, Onkeyup:**
> Queremos registrar as diferentes etapas quando uma tecla é pressionada no campo de entrada.\
>**Passo a passo:**
>1.  Selecione o campo de entrada usando document.getElementById.
>2.  Adicione eventos (onkeydown, onkeypress, onkeyup) ao campo.
>3.  No manipulador de eventos, exiba mensagens em um parágrafo na tela para cada etapa.

# **5. Onmouseover e Onmouseout:**
> Mudar a cor de fundo quando o mouse está sobre o elemento "box".\
>**Passo a passo:**
>1.  Selecione o elemento "box" usando document.getElementById.
>2.  Adicione eventos (onmouseover e onmouseout) ao elemento.
>3.  No manipulador de eventos, mude a cor de fundo ao passar o mouse e retorne à cor original ao retirar o mouse.

# **6. Ondrop, Ondragover, Ondragstart:**
>Aprimorar a interatividade do elemento "box" ao implementar a funcionalidade de arrastar e soltar. Isso permitirá que você mova o elemento pela página de forma dinâmica.\
>**Passo a passo:**
>1.  **Selecione o Elemento "box":**
>>Utilize document.getElementById para obter uma referência ao elemento com o id "box".
>2.  **Adicione o Evento Ondragstart para Iniciar o Arrasto:**
>>Adicione o evento ondragstart ao elemento "box".
Esse evento é acionado quando você inicia o processo de arrastar o elemento.
>3.  **Adicione o Evento Ondragover para Permitir a Sobreposição:**
>>Adicione o evento ondragover ao elemento "box".
Este evento é acionado enquanto o elemento é arrastado sobre uma área válida.
3.  **Adicione o Evento Ondrop para Lidar com o Evento de Soltar:**
>>Adicione o evento ondrop ao elemento "box". Esse evento é acionado quando você solta o elemento.

# **7. Onload:**
> Exibir um alerta quando a página estiver completamente carregada.\
>**Passo a passo:**
>1.  Selecione a janela usando window.
>2.  Adicione um evento de carregamento (onload) à janela.
>3.  No manipulador de eventos, exiba um alerta.\
Exemplo: `window.onload`

# **Desafio Extra: Jogo de "Clique-me!" com Contador**
>Desenvolver um jogo simples onde o desafio é clicar no botão "Desafio Extra" o máximo de vezes possível em um período de 10 segundos. O número de cliques deve ser exibido ao final do tempo.

> Dica: utilize o "setTimeout" para o tempo: https://www.w3schools.com/jsref/met_win_settimeout.asp

>**Instruções:**
>1.  **Botão "Desafio Extra":**
>>Foi adicionado um botão chamado "Desafio Extra" no seu projeto. Seu objetivo é clicar neste botão o máximo de vezes possível.
>2.  **Contador de Cliques:**
>>Um contador de cliques foi incluído abaixo do botão.
Cada vez que você clicar no botão, o contador será incrementado.
>3.  **Tempo Limite:**
>>Você terá um total de 10 segundos para realizar cliques.
>4.  **Ao Final do Tempo:**
>>Após o tempo de 10 segundos, um alerta será exibido informando quantos cliques você conseguiu.
